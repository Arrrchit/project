# IMPORT LIBRARIES


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Read CSV FILE

```python
s=pd.read_csv('https://api.covid19india.org/csv/latest/state_wise.csv')
```

```python
s
```

# INSPECT DATAFRAME

```python
s.info()
```

# COVID-19 DATA STATE WISE

```python
data=s.groupby('State').sum()
data
```

```python
data.drop(index=['Total','State Unassigned'],inplace=True)
```

```python
data.drop(columns=['Delta_Confirmed','Delta_Recovered','Delta_Deaths'],axis=1,inplace=True)
```

```python
data
```

```python
data.plot.bar(figsize=(20,8),width=0.8)
plt.show()
```

# SHOWING DATA IN GRADIENT COLORS

```python
data.style.background_gradient(cmap='Reds')
```

# TOP 10 STATES HAVING MOST NUMBER OF ACTIVE CASES

```python
v=data.groupby('State')['Active'].sum().sort_values(ascending=False).to_frame()
```

```python
v.head(10).style.background_gradient(cmap='Blues')
```

```python
v.head(10).plot.bar(figsize=(20,8))
plt.show()
```

# STATES HAVING MOST NUMBER OF DEATH CASES

```python
r=data.groupby('State')['Deaths'].sum().sort_values(ascending=False).to_frame()
r.head(10).style.background_gradient(cmap='Reds')
```

```python
r.head(10).plot.bar(figsize=(20,8),color='red')
plt.show()
```

# STATES HAVING MOST NUMBER OF RECOVERED CASES

```python
e=data.groupby('State')['Recovered'].sum().sort_values(ascending=False).to_frame()
e.head(10).style.background_gradient(cmap='Greens')
```

```python
e.head(10).plot.bar(figsize=(20,8),color='green')
plt.show()
```

# TOTAL NUMBER OF DEATHS TILL NOW

```python
data['Deaths'].sum()
```

# LIST OF STATES HAVING ACTIVE CASES LESS THAN 1000

```python
w=data[data.Active<1000]
w['Active'].sort_values(ascending=True).to_frame()
```

```python
w.plot.bar(figsize=(20,8))
plt.plot()
```

# States having minimum number of deaths

```python
e=data.groupby('State')['Deaths'].sum().sort_values(ascending=True).to_frame()
```

```python
e.head(15)
```

# TO ADD AN EXTRA COLUMN: "RECOVERY RATIO"

```python
for ind,row in data.iterrows():
    data.loc[ind,"Recovery Ratio"]=row['Recovered']/row['Confirmed']*100
```
